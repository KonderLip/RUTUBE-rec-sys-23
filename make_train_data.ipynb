```python
import numpy as np
import pandas as pd
from tqdm.auto import tqdm
```

```python
train = pd.read_parquet('./data/player_starts_train.parquet').sort_values('date')
train
```

```python
counts = train.user_id.value_counts()
counts
```

```python
np.mean(counts)
```

```python
users_to_drop = counts[counts<=6].index
```

```python
data = {}
```

```python
for user_id, item_id in tqdm(zip(train.user_id, train.item_id), total=len(train)):
    if user_id in users_to_drop:
        continue
    if user_id not in data:
        data[user_id] = []
    data[user_id].append(item_id)
```

```python
ids = []
with open('./data/train', 'w') as f:
    for user_id, line in data.items():
        ids.append(user_id)
        for value in line[-256:]:
            f.write(value.split('_')[1])
            f.write(' ')
        f.write('\n')
np.save('./data/train_user_ids.npy', ids)
```
